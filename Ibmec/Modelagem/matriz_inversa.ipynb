```python
import numpy.linalg as npl
import numpy as np

A = np.array(([0.52, 0.30, 0.18], [0.20, 0.50, 0.30], [0.25, 0.20, 0.55]))

b= np.array(([4800], [5810], [5690]))

I = npl.inv(A)

print("Matriz invertida ")
print(I)
x = np.dot(I,b)
print("O engenheiro deve retirar as seguintes quantidades da Mina 1,2,3 respectivamente: \n", x)

#Respostas

novob1 = ((x[1]+1000)- I[1,1]*b[1]-I[1,2]*b[2])/I[1,0]

variacao = novob1/b[0]-1
print("\nSe retirarmos 1000 m3 a mais da mina 1, a nova quantidade de areia será: \n")
print(novob1)
print("\nO percentual de variacao foi de: \n")
print(variacao)

novob11 = ((x[0] + 1000) - I[0,1]*b[1] - I[0,2]*b[2])/I[0,0]

variacao2 = novob11/b[0]-1

print("\nSe retirarmos 1000 m3 a mais da mina 2, a nova quantidade de areia será: \n")
print(novob11)
print("\nO percentual de variacao foi de: \n")
print(variacao2)
```
