```python
import numpy.linalg as npl
import numpy as np

# Linhas = materiais (areia, cascalho fino, cascalho grosso)
# Colunas = minas (M1, M2, M3)
A = np.array([[0.52, 0.20, 0.25],
              [0.30, 0.50, 0.20],
              [0.18, 0.30, 0.55]])

b = np.array([[4800], [5810], [5690]])

I = npl.inv(A)

print("Matriz inversa:")
print(I)

y = np.dot(I, b)
print("\nO engenheiro deve retirar das Minas 1, 2 e 3, respectivamente (m³):")
print(y)

# --- Caso 1: Mina 2 extrai 1000 m³ a mais ---
# y[1] = I[1,0]*b[0] + I[1,1]*b[1] + I[1,2]*b[2]
# Queremos y[1] + 1000, variando só b[0] (areia)
novob1 = ((y[1] + 1000) - I[1,1]*b[1] - I[1,2]*b[2]) / I[1,0]
variacao1 = novob1 / b[0] - 1

print("\nSe retirarmos 1000 m³ a mais da Mina 2, a nova quantidade de areia será:")
print(novob1)
print("O percentual de variação foi de:")
print(variacao1)

# --- Caso 2: Mina 1 extrai 1000 m³ a mais ---
# y[0] = I[0,0]*b[0] + I[0,1]*b[1] + I[0,2]*b[2]
# Queremos y[0] + 1000, variando só b[0] (areia)
novob2 = ((y[0] + 1000) - I[0,1]*b[1] - I[0,2]*b[2]) / I[0,0]
variacao2 = novob2 / b[0] - 1

print("\nSe retirarmos 1000 m³ a mais da Mina 1, a nova quantidade de areia será:")
print(novob2)
print("O percentual de variação foi de:")
print(variacao2)
```
