```python
#ac1
import matplotlib.pyplot as plt
import numpy as np
import math

print("Calculadora Série de Taylor: Cos(x)")

ntermos = int(input("Quantos termos essa Série de Taylor terá ?"))
xinicial = float(input("Qual o valor do X inicial ?"))
xfinal = float(input("Qual o valor do X final ?"))

xdiferenca = xfinal - xinicial

soma_taylor = 0

for n in range(ntermos):
    resto = n % 4

    if resto == 0:
        derivada = np.cos(xinicial)
    elif resto == 1:
        derivada = -np.sin(xinicial)
    elif resto == 2:
        derivada = -np.cos(xinicial)
    elif resto == 3:
        derivada = np.sin(xinicial)

    termo = (derivada * (xdiferenca**n))/math.factorial(n)

    soma_taylor += termo

valor_real = np.cos(xfinal)

erro = (valor_real - soma_taylor)/valor_real
valor_base_y = np.cos(xinicial)


print("-" * 30)
print(f"O valor da função (y) no xinicial={xinicial} é: {valor_base_y:.4f}")
print(f"O valor da função (y) no xfinal={xfinal} é: {valor_real:.4f}")
print(f"O valor aproximado da ordem {ntermos-1} é: {soma_taylor:.4f}")
print(f"Erro Relativo: {erro*100:.2f}%")

x_range = np.linspace(xinicial-1, xfinal+1) 
y_range = np.cos(x_range)

plt.plot(x_range, y_range, 'b-', label='f(x) = cos(x)')
plt.plot(xinicial, valor_base_y, 'ro', label='Ponto inicial')
plt.plot(xfinal, valor_real, 'bd', label='Ponto final')
plt.plot(xfinal, soma_taylor, 'kx', label=f'Taylor ordem {ntermos-1}')

plt.title(f"Taylor cos(x) - Aproximação com {ntermos} termos")
plt.legend()   
plt.grid() 
plt.show()
```
