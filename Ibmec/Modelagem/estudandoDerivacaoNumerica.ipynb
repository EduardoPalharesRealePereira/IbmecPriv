# EX 1 - treinar

```python
import numpy as np
import matplotlib.pyplot as plt

# Função e derivadas teóricas
f  = lambda x: x**3 - 2*x + 4
f1 = lambda x: 3*x**2 - 2          # 1ª derivada teórica
f2 = lambda x: 6*x                  # 2ª derivada teórica

x = np.arange(-2, 2 + 0.25, 0.25)
h = 0.25

# Diferenças finitas — 1ª derivada
prog1  = (f(x + h) - f(x)) / h
regr1  = (f(x) - f(x - h)) / h
cent1  = (f(x + h) - f(x - h)) / (2 * h)

# Diferenças finitas — 2ª derivada
prog2  = (f(x + 2*h) - 2*f(x + h) + f(x)) / h**2
regr2  = (f(x) - 2*f(x - h) + f(x - 2*h)) / h**2
cent2  = (f(x + h) - 2*f(x) + f(x - h)) / h**2

# Gráficos
# Gráfico da 1ª Derivada
plt.figure(figsize=(13, 5))

plt.subplot(1, 2, 1)
plt.plot(x, prog1, "--o", label="Progressiva", markersize=4)
plt.plot(x, regr1, "--s", label="Regressiva",  markersize=4)
plt.plot(x, cent1, "--^", label="Centrada",    markersize=4)
plt.plot(x, f1(x), "-k",  label="Teórica")
plt.title("1ª Derivada")
plt.legend()
plt.grid(True)

# Gráfico da 2ª Derivada
plt.subplot(1, 2, 2)
plt.plot(x, prog2, "--o", label="Progressiva", markersize=4)
plt.plot(x, regr2, "--s", label="Regressiva",  markersize=4)
plt.plot(x, cent2, "--^", label="Centrada",    markersize=4)
plt.plot(x, f2(x), "-k",  label="Teórica")
plt.title("2ª Derivada")
plt.legend()
plt.grid(True)

plt.suptitle("Diferenças Finitas — $f(x)=x^3-2x+4$", fontsize=14)
plt.tight_layout()
plt.show()

```
