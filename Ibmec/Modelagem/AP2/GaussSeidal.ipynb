```python

import numpy as np
import scipy.sparse.linalg as sp

#metodo de gradiente conjugado chamado cgs da bibilioteca scipy.sparse.linalg
#nao vamos ter que implementar gauss seidal na mao, apenas no cgs
A = np.array([[10,2,-1], [-3,-6,2], [1,1,5]])
b = np.array([[27], [-61.5], [-21.5]])
x = sp.cgs(A, b, [0,0,0], maxiter=4)

print("O resultado do sistema é: ", x[0])
print(x)
```

Gauss seidal e jacobi sao a mesma coisa na pratica. A diferença é que Jacobi começa no ponto (0,0,0) e Gauss Seidal ja vai substituindo os pontos que for obtendo na propria interação.
