```python
from numpy import linspace
from scipy.integrate import trapezoid

def f(x):
 y = 0.2+25*x-200*x**2+675*x**3-900*x**4+400*x**5 #perguntar porque são 11 pontos no linspace de baixo se sao apenas 6 pontos na função
 return y

x = linspace(0,0.8,11)
I = trapezoid(f(x), x)
erro = abs((1.640533-I)/1.650533)
print("Valor de I: ", I)
print(f"O erro relativo em relação ao valor exato é: {round(erro*100,2)} %")
```
