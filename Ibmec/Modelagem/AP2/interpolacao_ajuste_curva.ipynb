Exemplo prototipo do slide


```python
from numpy import array, polyfit, poly1d, log10
from matplotlib.pyplot import plot, xlabel, ylabel, legend, title

b = array([0.9030900, 0.9542425, 1.0413927, 1.0791812])
x = array([8, 9, 11, 12])  # não é necessário ordenar os pontos
A = polyfit(x, b, 3)  # cálculo dos coeficientes
f = poly1d(A)  # A função poly1d cria um polinômio a partir dos coeficientes passados em A
erro = abs((log10(10) - f(10)) / log10(10)) * 100

plot(x, b, "kd", label="Pontos base")
plot(x, f(x), "r", label="Polinômio de ordem 3")
title("Ajuste Polinomial")
xlabel("x")
ylabel("y")
legend()

print("O polinomio ajustado é:\n", f )
print("\nO valor ajustado para x = 10 é: ", f(10))
print(f"O erro relativo percentual é: {round(erro, 5)}%")
```
