```python
import numpy as np
import matplotlib.pyplot as mlb
def f(x):
    y = 1/(1+25*x**2)
    return y
x = np.linspace(-1,1,21)
b = f(x)
c = np.polyfit(x,b,6)
p = np.poly1d(c)
mlb.plot(x, f(x), 'kd', label = 'Dados tabelados')
mlb.plot(x, p(x), 'r', label = 'Polinomio')
mlb.title('Ajuste polinomial')
mlb.xlabel('x')
mlb.ylabel('f(x)')
mlb.legend()
```
