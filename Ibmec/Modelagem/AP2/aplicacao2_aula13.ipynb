```python
import numpy as np
import matplotlib.pyplot as mlb
def f(x):
    y = 1/(1+25*x**2)
    return y
x = np.linspace(-1,1,21)
b = f(x)
c = np.polyfit(x,b,6)
p = np.poly1d(c)
mlb.plot(x, f(x), 'kd', label = 'Dados tabelados')
mlb.plot(x, p(x), 'r', label = 'Polinomio')
mlb.title('Ajuste polinomial')
mlb.xlabel('x')
mlb.ylabel('f(x)')
mlb.legend()
```

```python
from scipy.interpolate import lagrange
from numpy import array, linspace
from matplotlib.pyplot import plot, xlabel, ylabel, legend, title

x = array([-1, 0, 3])
y = array([15, 8, -1])
f = lagrange(x, y)
xx = linspace(-1, 3, 50)
plot(xx, f(xx), 'g', label='Função Interpolada')
plot(x, y, 'ro', label='Dados Tabelados')
plot(1, f(1), '*b', label='Interpolação de Ordem 2 para f(1)')
title('Ajuste Polinomial de Lagrange')
xlabel('x')
ylabel('f(x)')
legend()
print("O polinomio ajustado é:\n", f)
print("\nO valor ajustado para x=1 é:", f(1))
```
