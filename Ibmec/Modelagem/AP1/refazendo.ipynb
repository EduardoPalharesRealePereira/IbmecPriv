```python

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import scipy.optimize
```

```python
f = lambda x: np.sin*(0.5*np.sqrt(x))/x
x = np.array([1,3,5,7])
y = f(x)
dy = np.gradient(y, x)

plt.plot(x, dy, label = 'x')
plt.plot(y, dy)
```

```python
x = sp.symbols("x")

f = 3*x**2 + x**2


df = sp.diff(f, x)
print(df)
```
