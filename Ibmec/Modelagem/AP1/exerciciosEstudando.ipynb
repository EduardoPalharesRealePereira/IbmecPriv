```python


import numpy as np  

tempo = np.array([0, 1, 2, 3])
volume = np.array([0, 1, 2.3, 3.9])

h = tempo[1] - tempo[0]

q = np.zeros(len(tempo))

## progressiva - primeiro ponto
q[0] =  (volume[1] - volume[0])/ h

## regressiva - ultimo ponto

q[3] = (volume[3] - volume[2])/h    

##centrada - 2 pontos do meio

q[1] = (volume[2] - volume [0])/(2*h)
q[2] = (volume[3] - volume[1])/(2*h)

print(f"Vazao = {q}")




```
